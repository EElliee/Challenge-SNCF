## MODELE OPTIMAL 
### SCORE=0.098

```python
xbest=meilleurX_train
xtbest=meilleurX_test
y=Y_train
```

```python
### Optimisation des hyperparametres avec l'optimisation  baysienne 

#### ATTENTION CODE LONG A EXECETUER ON A LANCER AVEC max_evals=2
#### POUR TESTER LE CODE MAIS LE RESULTAT QU'ON DETAILLE SUR LE RAPPORT EST AVEC 100

import numpy as np
from sklearn.model_selection import cross_val_score,KFold
from sklearn.ensemble import RandomForestRegressor
from hyperopt import fmin, tpe, hp, Trials,rand


# Dfinir la fonction objectif pour l'optimisation
def objective(params):
   ### Convertir max_features en valeur absolue si ncessaire
    if params['max_features'] == 'sqrt':
        params['max_features'] = int(np.sqrt(xbest.shape[1]))
    elif params['max_features'] == 'log2':
        params['max_features'] = int(np.log2(xbest.shape[1]))
    elif params['max_features'] is None:
        # Remplacez la valeur par le nombre total de variables
        params['max_features'] = xbest.shape[1]

   #### Crer le modle RandomForestRegressor avec les hyperparamtres  valuer
    model = ExtraTreesRegressor(
        n_estimators=int(params['n_estimators']),
        max_depth=int(params['max_depth']),
        min_samples_split=int(params['min_samples_split']),
        min_samples_leaf=int(params['min_samples_leaf']),
        max_features=int(params['max_features']),
        
        #### bootstrap=params['bootstrap'],
        random_state=42
    )

    myfold=KFold(n_splits=10)
    # Calculer la performance en utilisant la validation croise avec MSE
    mse_scores = -cross_val_score(model, xbest, np.array(y).ravel(), cv=myfold, scoring='neg_mean_absolute_error')

    ### Retourner la moyenne des scores MSE (la fonction objectif doit tre minimise)
    return np.mean(mse_scores)

#### Dfinir l'espace d'hyperparamtres
param_space = {
    'n_estimators': hp.quniform('n_estimators', 100, 1000, 10),
    'max_depth': hp.quniform('max_depth', 10, 200, 10),
    'min_samples_split': hp.quniform('min_samples_split', 2, 20, 1),
    'min_samples_leaf': hp.quniform('min_samples_leaf', 1, 20, 1),
    'max_features': hp.choice('max_features', [int(x) for x in np.arange(1, 13)] + ['sqrt', 'log2', None]),
    #'bootstrap': hp.choice('bootstrap', [True, False])
}

random_state = np.random.RandomState(42)
trials = Trials()
best_params_hyperopt = fmin(fn=objective, space=param_space, algo=rand.suggest, max_evals=100,trials=trials)

### Afficher les meilleurs hyperparamtres
print("Meilleurs hyperparamtres (optimisation baysienne avec Hyperopt):", best_params_hyperopt)

#### Crer le modle RandomForestRegressor avec les meilleurs hyperparamtres
best_model_hyperopt = ExtraTreesRegressor(
    n_estimators=int(best_params_hyperopt['n_estimators']),
    max_depth=int(best_params_hyperopt['max_depth']),
    min_samples_split=int(best_params_hyperopt['min_samples_split']),
    min_samples_leaf=int(best_params_hyperopt['min_samples_leaf']),
    max_features=int(best_params_hyperopt['max_features']),
    
    #bootstrap=bool(best_params_hyperopt['bootstrap']),
    random_state=42
)

####Entraner le meilleur modle sur l'ensemble de donnes complet
best_model_hyperopt.fit(xbest, np.array(y).ravel())

# Faire des prdictions sur de nouvelles donnes
prediction = best_model_hyperopt.predict(xtbest)
```
